### Definindo função para remover caracteres especiais

```python
def remover_acentos(txt):
    return normalize('NFKD', txt).encode('ASCII','ignore').decode('ASCII')
```

## Formatando o arquivo para a leitura

```python
with open(OriginalFileName, 'r') as file :
  filedata = file.read()

filedata = filedata.replace(',', '.')
filedata = remover_acentos(filedata)

with open(NewFileName, 'w') as file:
  file.write(filedata)
```

## Lendo arquivo

```python
data = pd.read_csv(NewFileName, sep=';')
```

```python
pd.to_numeric(data.vl_despesa, errors='coerce')
df = pd.DataFrame(data)
#df.set_index('id_despesa_detalhe', inplace=True)
#data.vl_despesa
```

## Percorrendo dados de cada grupo

```python
for grpName in uniques:
    elem = df[df.ds_acao == grpName]
    dfElem = pd.DataFrame(elem)
    dfElem.describe()
    
    elem.plot(x=elem.index, y='vl_despesa', kind='line', title=grpName, figsize=(10,5))
    
    
```
